```python
import pandas as pd
```

```python
df = pd.read_csv("ptet_cnp_provinces.csv", delimiter=";")
df.head()
```

```python
df = df.rename(columns={
    "Province": "Région"
})
```

```python
df_melted = df.melt(id_vars=["Région", "Profession"],
        var_name="Date",
        value_name="PTÉT")

df_melted["PTÉT"] = pd.to_numeric(df_melted["PTÉT"], errors="coerce")
df_melted.head()
```

```python
df_melted = df_melted.set_index("Date")
df_melted.head()
```

```python
df_melted = df_melted[df_melted["PTÉT"] != 0]
df_melted = df_melted.dropna()
df_melted["PTÉT"] = df_melted["PTÉT"].fillna(0).astype(int)
df_melted.head()
df_melted
```

```python
df_melted[df_melted["Région"] == "Canada"]["PTÉT"].isna().sum()
```

```python
df_melted.to_csv("ptét_cnp_analyse.csv")
```
